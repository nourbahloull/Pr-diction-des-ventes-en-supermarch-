```python
from scipy import stats
quantile = []
for i in range (1,100):
  v=stats.norm.ppf(i/100)
  quantile.append(v)

  print("quantile",i,quantile)

```

```python
diff=[]
for i in range(0,49) :
  d=quantile[i+1]-quantile[i]
  diff.append(d)
  print("differnece",i+1,diff)
```

```python
dif1=[]
for i in range(1,50) :
  d=quantile[i+1]-quantile[i]
  dif1.append(d)
  print("differnece",i+1,dif1)
```

```python
for j in range(len(diff)-1,-1,-1):
    if(diff[j]!=dif1[i]):
      print("non sym","Q",j,"Q",i,"differnce=")
      break;
else:
    print("symetrique")
```
